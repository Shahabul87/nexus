## Kramer Zone Analysis

Jaundice severity is assessed using Kramer zones:
- Zone 1: Face only (mild)
- Zone 2: Chest (moderate)
- Zone 3: Abdomen (moderate-severe)
- Zone 4: Arms/legs (severe)
- Zone 5: Palms/soles (critical)
